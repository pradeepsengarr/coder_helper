```python
import gradio as gr
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
MODEL_NAME = "google/gemma-2b-it"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, torch_dtype=torch.float16, device_map="auto")
```

```python
SYSTEM_PROMPT = """
You are an expert coding assistant skilled in Python, JavaScript, and C++.
Provide well-structured, optimized, and efficient code solutions with explanations.
Follow best practices, include comments, and ensure the code is production-ready.
"""
```

```python
def generate_code(user_input):
    prompt = f"{SYSTEM_PROMPT}\nUser: {user_input}\nAI:"
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    output = model.generate(**inputs, max_length=512, temperature=0.7, top_p=0.9)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response
```

```python
demo = gr.Interface(
    fn=generate_code,
    inputs=gr.Textbox(lines=3, placeholder="Enter your coding question here..."),
    outputs=gr.Textbox(),
    title="Expert Coding Assistant",
    description="Ask any coding-related question, and get expert-level responses with optimized code!"
)
```

```python
demo.launch()
```
